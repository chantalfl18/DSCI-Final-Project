### 1) Data Description

The **`players`** features **196** observations with **7** variables. 
    
Variables:
+ **`experience`**: character variable, five categories for the experience level of the player: 'Pro', 'Veteran', 'Amateur', 'Regular', 'Beginner'.
+ **`subscribe`**: logical variable, two categories for if the player is subscribed or not, 'True', 'False'.
+ **`hashedEmail`**: character variable, records the hashed email of the player
+ **`played_hours`**: double variable, records the hours logged by the player, hours from 0 to 223.1.
+ **`name`**: character variable, records the name of the players
+ **`gender`**: character variable, seven categories of the gender of the player, 'Male', 'Female', 'Non-binary', 'Prefer not to say', 'Agender', 'Two-Spirited', 'Other'
+ **`Age`**: double variable, records the age of the player, ages from 9 to 58.


The **`sessions`** features **1535** observations with **5** variables. 

Variables:
+ **`hashedEmail`**: character variable, records the hashed email of the player
+ **`start_time`**: character variable, records the start date and time of the player
+ **`end_time`**: character variable, records the end date and time of the player
+ **`original_start_time`**: double variable, records the original start time of the player, from 1.71e+12 to 1.73e+12.
+ **`original_end_time`**: double variable, records the original end time of the player, from 1.71e+12 to 1.73e+12.

#### Potential Issues:

### 4) Methods and Plan

KNN regression will be the predictive model used to answer the question mentioned above. This is because I am attempting to predict a numerical value. The use of KNN regression assumes that the data is not linear and is properly scaled. Furthermore, KNN regression may be easily influenced by outliers, thereby skewing prediction accuracy. To compare different models, cross-validation is required to determine the optimal k value, and evaluation metrics such as RMSE can be used to compare the accuracy of KNN regression with linear regression. To process the dataset, I would first need to standardize the data. Because `experience` is categorical, I would either have to enumerate the categories or standardize the data in a way that the categorical values of `experience` are on a comparable scale to the values of `Age`. The training and test proportions will be 60/40. I will then use 5-fold cross-validation to tune the parameter. After selecting the optimal predictor model, I will retrain the model using the training set and then evaluate the accuracy with the testing set.
